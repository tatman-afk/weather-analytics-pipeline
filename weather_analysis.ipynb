```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.plot(
    df["day"],
    df["avg_temp"],
    marker="o",
    linewidth=2,
    label="Average Temperature"
)

plt.fill_between(
    df["day"],
    df["min_temp"],
    df["max_temp"],
    alpha=0.3,
    label="Daily MinMax Range"
)

plt.title("Daily Temperature Summary (Tucson, AZ)")
plt.xlabel("Date")
plt.ylabel("Temperature (F)")
plt.legend()
plt.grid(True)

plt.show()
```
